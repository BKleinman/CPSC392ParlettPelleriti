```python
penguins = pd.read_csv("https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/penguins.csv")
```

```python
print(penguins.info())
```

```python
indx = (penguins["sex"] == "male") | (penguins["sex"] == "female")
penguins2 = penguins.loc[indx]
```

```python
(ggplot(penguins2, aes(x = "body_mass_g", y = "flipper_length_mm", fill = "sex")) + geom_point() + facet_wrap("~sex") 
 + stat_smooth(method = "lm", se = True) + theme_minimal() 
 + labs(x = "Body Mass (Grams)", y = "Flipper Length (mm)")
 + ggtitle("Body Mass Vs. Flipper Length in Penguins")
 + scale_fill_manual(["#cc79a7", "#0072b2"]))
```
