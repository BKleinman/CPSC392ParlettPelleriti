```python
fifa = pd.read_csv("https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/players_15.csv")

fifa["monthBorn"] = fifa["dob"].str.extract(r'-([0-9][0-9])-')

fifa.iloc[1:30,9]

```

```python
(ggplot(fifa, aes(x = "age")) + geom_histogram())
```

```python
(ggplot(fifa, aes(x = "preferred_foot", y = "weight_kg")) + geom_boxplot())
```

```python
clubs = (fifa["club"] == "Chelsea") | (fifa["club"] == "Manchester United")
(ggplot(fifa.loc[clubs], aes(x = "club", y = "height_cm")) + geom_boxplot())
```

```python
fifa["monthBorn"] = fifa["dob"].str.extract(r'-([0-9][0-9])-')
```

```python
(ggplot(fifa, aes(x = "weight_kg", y = "height_cm")) + geom_point() + stat_smooth(method = "lm", color = "red") + facet_wrap("~monthBorn"))
```

```python
(ggplot(fifa, aes(x = "weight_kg", y = "height_cm")) + geom_point() + stat_smooth(method = "lm", color = "red") + facet_wrap("~body_type"))
```
