```python
penguin_url = 'https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/penguins.csv'
titanic_url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv'

penguins = pd.read_csv(penguin_url)
titanic = pd.read_csv(titanic_url)

print(penguins.head())
print("-------------")
print(titanic.head())
print("-------------")

sex = titanic["sex"]
print(sex)
print("-------------")

penguins_10 = penguins.iloc[[10]]
print(penguins_10)
print("-------------")

passenger12 = titanic.loc[12, "age"]
print(passenger12)
print("-------------")

mean_age = titanic["age"].mean()
print(mean_age)
print("-------------")

penguins["bill_ratio"] = penguins["bill_length_mm"]/penguins["bill_depth_mm"]
print(penguins["bill_ratio"])
print("-------------")

indx = penguins["body_mass_g"] < 4000
lil_boys = penguins.loc[indx]
print(lil_boys.iloc[0:4])
print("-------------")

bl_Adelie = penguins.loc[penguins["species"] == "Adelie", "bill_length_mm"].mean()
bl_Chinstrap = penguins.loc[penguins["species"] == "Chinstrap", "bill_length_mm"].mean()
bl_Gentoo = penguins.loc[penguins["species"] == "Gentoo", "bill_length_mm"].mean()
print("Mean Adelie Bill Length: " + str(bl_Adelie))
print("Mean Chinstrap Bill Length: " + str(bl_Chinstrap))
print("Mean Gentoo Bill Length: " + str(bl_Gentoo))
print("-------------")

class1_prop = titanic.loc[titanic["pclass"] == 1, "survived"].sum()/len(titanic.loc[titanic["pclass"] == 1])
class2_prop = titanic.loc[titanic["pclass"] == 2, "survived"].sum()/len(titanic.loc[titanic["pclass"] == 2])
class3_prop = titanic.loc[titanic["pclass"] == 3, "survived"].sum()/len(titanic.loc[titanic["pclass"] == 3])
print("Class 1 survivor proportion: " + str(class1_prop))
print("Class 2 survivor proportion: " + str(class2_prop))
print("Class 3 survivor proportion: " + str(class3_prop))
print("-------------")

meanMaleFare = titanic.loc[titanic["sex"] == "male", "fare"].mean()
meanFemaleFare = titanic.loc[titanic["sex"] == "female", "fare"].mean()
print("Mean male fare was: " + str(meanMaleFare) + '\n' + "Mean female fare was: "+ str(meanFemaleFare))
print("-------------")

small_penguin = penguins[["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]]
print(small_penguin.head())
print("-------------")
```
